```python
import pandas as pd
```

```python
df = pd.read_csv("heart.csv")
```

```python
X = df.drop(["output","fbs","restecg"],axis=1)
y = df["output"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# https://github.com/dmlc/xgboost/issues/1725#issuecomment-295059572
import matplotlib.pyplot as plt
import xgboost as xgb

model = xgb.XGBClassifier()

model.fit(X_train,y_train)

xgb.plot_tree(model)
fig = plt.gcf() # get current figure
fig.set_size_inches(200, 200) # deepending on the length of the tree customize size
plt.show()

# fig.savefig('tree.png') # to save as png
```
